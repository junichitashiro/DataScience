# 線形回帰

```python
import pandas as pd
```

## データの確認

### データの読み込み

```python
df = pd.read_csv('sample_data/rabbit.csv')
df
```

### 相関係数の分析

```python
df.corr()
```

## データの視覚化

```python
import matplotlib.pyplot as plt
```

### 散布図の表示

```python
plt.scatter(df['大きさ'], df['重さ'])
```

## モデルの作成

### 目的変数と説明変数の設定
y:目的変数、x:説明変数

```python
y = df['重さ']
x = df['大きさ'].to_frame()
```

### 線形回帰モデルの作成

```python
import numpy as np
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
x_array = np.array(x)
lr.fit(x_array, y)
```

### グラフ化して確認

```python
plt.plot(x_array, lr.predict(x_array))
```

## 予測の実行

### 説明変数を手動で入力して予測を実行

```python
prediction = lr.predict(np.array([[25]]))
print(prediction)
```
