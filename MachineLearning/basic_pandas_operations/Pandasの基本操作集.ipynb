## 6.条件を指定して抽出する

* heroes.csvを読み込みdf_heroesで定義する
* 以下の条件で抽出する
  * 種族 が Human
  * 身長 が170以上200未満


```python
df_heroes = pd.read_csv('heroes.csv')
```

```python
df_heroes[df_heroes['種族'] == 'Human']
```

```python
df_heroes.query('種族 == "Human"')
```

```python
df_heroes[df_heroes['種族'].isin(['Human'])]
```

```python
df_heroes[(df_heroes['身長'] >= 175) & (df_heroes['身長'] < 200)]
```

```python
df_heroes.query('身長 >= 175 & 身長 < 200')
```
