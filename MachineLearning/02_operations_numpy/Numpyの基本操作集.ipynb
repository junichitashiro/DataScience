## 7.四則演算

足し算

$
  \begin{pmatrix}
     1 & 3\\
    -2 & 4\\
  \end{pmatrix}
  +\begin{pmatrix}
    2 & -1\\
    3 &  0\\
  \end{pmatrix}
$

```python
a = np.array([[1, 3], [-2, 4]])
b = np.array([[2, -1], [3, 0]])
a + b
```

引き算

$
  \begin{pmatrix}
     1 & 3\\
    -2 & 4\\
  \end{pmatrix}
  -\begin{pmatrix}
    2 & -1\\
    3 &  0\\
  \end{pmatrix}
$

```python
a = np.array([[1, 3], [-2, 4]])
b = np.array([[2, -1], [3, 0]])
a - b
```

行列積

$
  \begin{pmatrix}
    -2 & -5\\
     1 &  3\\
  \end{pmatrix}
  \begin{pmatrix}
    3 &   0\\
    6 &  -2\\
  \end{pmatrix}
$

```python
a = np.array([[-2, -5], [1, 3]])
b = np.array([[3, 0], [6, -2]])
a @ b
```

```python
a.dot(b)
```

別解

アダマール積

$
  \begin{pmatrix}
     3 &  5\\
     4 & -1\\
  \end{pmatrix}
  \cdot\begin{pmatrix}
    -2 &  1\\
     0 & -3\\
  \end{pmatrix}
  =\begin{pmatrix}
    3*(-2) &     5*1\\
       4*0 & -1*(-3)\\
  \end{pmatrix}
$

```python
a = np.array([[3, 5], [4, -1]])
b = np.array([[-2, 1], [0, -3]])
a * b
```

## 9.逆行列

### 対象となる行列

$
  \boldsymbol{A} =
    \begin{pmatrix}
       4 & -2 \\
      1 &  0 \\
    \end{pmatrix}
$

### 取得する値

* 行列の行列式の値
* 行列式の値が0でなければその逆行列

```python
A = np.array([[4, -2], [1, 0]])
A
```

linalg = linear algebra（線形代数）

det = determinant

```python
det = np.linalg.det(A)
det
```

```python
np.linalg.inv(A)
```

```python
Ainv = np.linalg.inv(A)
```

単位行列が1になっているかの検算

```python
A @ Ainv
```

## 15.行列を結合する

行列を縦方向、横方向にそれぞれ結合する

$
  \boldsymbol{A} =
    \begin{pmatrix}
      0 &  1 & -1\\
      2 &  4 & -3\\
      5 & -2 &  7\\
    \end{pmatrix}
$

$
  \boldsymbol{B} =
    \begin{pmatrix}
      -2 &  0 & 1\\
       5 & -1 & 2\\
      -6 &  3 & 4\\
    \end{pmatrix}
$


```python
A = np.array([0, 1, -1, 2, 4, -3, 5, -2, 7]).reshape(3, -1)
A
```

```python
B = np.array([-2, 0, 1, 5, -1, 2, -6, 3, 4]).reshape(3, -1)
B
```

### 縦方向に結合する

```python
C = np.vstack((A, B))
C
```

### 横方向に結合する

```python
D = np.hstack((A, B))
D
```
